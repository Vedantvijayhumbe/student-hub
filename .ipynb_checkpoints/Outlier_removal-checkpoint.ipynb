```python
import pandas as pd

df_wine = pd.read_csv('wines_SPA.csv')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
     
```

```python
num_cols = df_wine.select_dtypes(include=[np.number]).columns    
plt.figure(figsize=(15, 8))
df_wine[num_cols].boxplot(rot=45)
plt.title("Boxplot of Numerical Features to Detect Outliers")                     
plt.show()
```

```python
import scipy.stats as stats
numerical_cols = ["rating", "num_reviews", "price", "body", "acidity"]
skewness_values = df_wine[numerical_cols].skew()
plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    
    #HISTOGRAM 
    sns.histplot(df_wine[col], bins=30, kde=True, stat="density", label="Actual Data", color="blue", alpha=0.6)
    
    # FITTING A NORMAL DISTRIBUTION TO THE DATA 
    mean, std = df_wine[col].mean(), df_wine[col].std()
    x = np.linspace(mean - 4*std, mean + 4*std, 100)
    y = stats.norm.pdf(x, mean, std)  # Normal distribution curve
    
    plt.plot(x, y, color="red", linestyle="dashed", label="Normal Curve")  # Overlay bell curve
    plt.axvline(mean, color='green', linestyle='dashed', linewidth=2, label="Mean")
    
    plt.title(f"{col} Distribution (Skewness: {skewness_values[col]:.2f})")
    plt.legend()



#reason for doing this stuff is seeing the skewness of the normal distribution curve , ans analysing z-score . 

plt.tight_layout()
plt.show()
```

```python
numerical_features = ['year', 'rating', 'num_reviews', 'price', 'body', 'acidity']

df_wine['year'] = pd.to_numeric(df_wine['year'], errors='coerce')

for col in numerical_features:

    if not pd.api.types.is_numeric_dtype(df_wine[col]):
        print(f"Column '{col}' is not numeric. Skipping z-score calculation.")
        continue

    # CALCULATING Z-SCORE 
    df_wine[f'{col}_zscore'] = stats.zscore(df_wine[col].dropna())

display(df_wine.head())
```

```python
outlier_methods = {}
zscore_threshold = 3
numerical_features = ['year', 'rating', 'num_reviews', 'price', 'body', 'acidity']

for col in numerical_features:
    # MOST IMPORTANT THING HERE IS IDENTIFYING OUTLIERS 
    outliers_positive = df_wine[df_wine[f'{col}_zscore'] > zscore_threshold]
    outliers_negative = df_wine[df_wine[f'{col}_zscore'] < -zscore_threshold]
    
    if len(outliers_positive) > 0 or len(outliers_negative) > 0:
      # THERE ARE 2 METHODS TO DEAL WITH OUTLIER REMOVAL - WE USE DATA CAPPING INCASE OF NORMAAL LIKE , ELSE TRIMMING IT 
      if col in ['rating', 'price', 'body', 'acidity']:
          # Capping
          percentile_95 = df_wine[col].quantile(0.95)
          percentile_05 = df_wine[col].quantile(0.05)
          df_wine.loc[df_wine[f'{col}_zscore'] > zscore_threshold, col] = percentile_95
          df_wine.loc[df_wine[f'{col}_zscore'] < -zscore_threshold, col] = percentile_05
          outlier_methods[col] = 'capping'
      else:
          # Trimming
          outliers = df_wine[(df_wine[f'{col}_zscore'] > zscore_threshold) | (df_wine[f'{col}_zscore'] < -zscore_threshold)].index
          df_wine.drop(outliers, inplace=True)
          outlier_methods[col] = 'trimming'

print(outlier_methods)
display(df_wine.head())
```

```python
import matplotlib.pyplot as plt

numerical_features = ['year', 'rating', 'num_reviews', 'price', 'body', 'acidity']

plt.figure(figsize=(20, 12))

for i, col in enumerate(numerical_features):
    plt.subplot(3, 2, i + 1)
    # Before outlier removal
    plt.boxplot(df_wine[col], positions=[1], widths=0.6, patch_artist=True, showfliers=True, tick_labels=['Before'])
    # After outlier removal
    plt.boxplot(df_wine[col], positions=[2], widths=0.6, patch_artist=True, showfliers=True, tick_labels=['After'])
    plt.title(f'Boxplot of {col} Before and After Outlier Treatment')
    plt.ylabel(col)
    plt.xticks([1, 2], ['Before', 'After'])
plt.tight_layout()
plt.show()
```
